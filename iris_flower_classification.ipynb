```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
iris.keys()
```

```python
print(iris.DESCR)
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
x=iris['data']
```

```python
x
```

```python
y=iris['target']
```

```python
x=pd.DataFrame(x,columns=['sepal length (cm)',
 'sepal width (cm)',
 'petal length (cm)',
 'petal width (cm)'])
```

```python
x.head()
```

```python
y
```

```python
sns.pairplot(x)
```

```python
x.corr()
```

```python
sns.heatmap(x.corr(),annot=True)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train.shape,y_train.shape)
```

```python
import warnings
warnings.filterwarnings(action='ignore')
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print(mean_squared_error(y_pred,y_test))
print(mean_absolute_error(y_pred,y_test))
print(r2_score(y_pred,y_test))
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
print(classification_report(y_pred,y_test))
print(accuracy_score(y_pred,y_test))
```

```python
print(confusion_matrix(y_pred,y_test))
```

```python
value=lr.predict([[4.6,3.1,1.5,2.0]])
print(f"The predicted specie is : {iris.target_names[value]}")
```

Decision Trees

```python
from sklearn.tree import DecisionTreeClassifier
trees=DecisionTreeClassifier()
trees.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(10,10))
tree.plot_tree(trees,filled=True)
```

```python
y_p=trees.predict(x_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
print(classification_report(y_p,y_test))
print(accuracy_score(y_p,y_test))
```
